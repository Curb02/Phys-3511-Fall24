```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

# Function to calculate exact ln(n!) using math.factorial()
def exact_ln_factorial(n):
    return math.log(math.factorial(n))

# Function to calculate Stirling's approximation for ln(n!)
def stirling_approx(n):
    return n * math.log(n) - n

# Arrays to store values of n, exact ln(n!), Stirling's approximation, and relative errors
n_values = np.arange(2, 101)  # Adjust range if needed for the error computation
exact_values = []
stirling_values = []
relative_errors = []

# Calculate exact values, approximations, and relative errors
for n in n_values:
    exact_ln = exact_ln_factorial(n)
    stirling_ln = stirling_approx(n)
    
    exact_values.append(exact_ln)
    stirling_values.append(stirling_ln)
    
    relative_error = 100 * abs(exact_ln - stirling_ln) / exact_ln
    relative_errors.append(relative_error)

# Plotting the exact ln(n!) and Stirling's approximation
plt.figure(figsize=(8, 6))
plt.plot(n_values, exact_values, label="Exact", color="blue")
plt.plot(n_values, stirling_values, label="Stirling's Approx.", linestyle='--', color="red")
plt.xlabel("n")
plt.ylabel("ln(n!)")
plt.legend()
plt.title("Exact vs Stirling's Approximation of ln(n!)")
plt.grid(True)
plt.show()

# Find the first n where the relative error is less than 1%
for n, error in zip(n_values, relative_errors):
    if error < 1:
        print(f"The first value of n where the relative error is less than 1% is n = {n}")
        break
```
