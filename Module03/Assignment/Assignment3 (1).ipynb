```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#part 1
k = 1
mp = 1
mc = 12 * mp
mo = 16 *mp

M1 = np.array([[k/mo,-k/mo,0],[-k/mc,2*k/mc,-k/mc],[0,-k/mo,k/mo]])
print(M1)
```

```python
#part 2

w,v = np.linalg.eig(M1)
w[2] = 0
print('E-value:', w)
print('E-vector', v)
```

```python
#part 3A
t = np.linspace(0,100,100)

x1 = v[0,0] * np.cos(np.sqrt(w[0])*t) 
x2 = v[1,0] * np.cos(np.sqrt(w[0])*t) 
x3 = v[2,0] * np.cos(np.sqrt(w[0])*t) 

fig1 = plt.figure()
ax = plt.axes()
ax.set_xlabel('t(s)')
plt.title('3A')
ax.plot(np.linspace(0,100,100),x1,color='blue', alpha=0.4)
ax.plot(np.linspace(0,100,100),x2,color='red', alpha=0.4)
ax.plot(np.linspace(0,100,100),x3,color='green', alpha=0.4)
```

```python
#part 3B

x1 = v[0,1] * np.cos(np.sqrt(w[1])*t) 
x2 = v[1,2] * np.cos(np.sqrt(w[1])*t) 
x3 = v[2,2] * np.cos(np.sqrt(w[1])*t) 

fig1 = plt.figure()
ax = plt.axes()
ax.set_xlabel('t(s)')
plt.title('3B')
ax.plot(np.linspace(0,100,100),x1,color='blue', alpha=0.4)
ax.plot(np.linspace(0,100,100),x2,color='red', alpha=0.4)
ax.plot(np.linspace(0,100,100),x3,color='green', alpha=0.4)
```

```python
#part 3C

x1 = v[0,2] * np.cos(np.sqrt(w[2])*t) 
x2 = v[1,2] * np.cos(np.sqrt(w[2])*t) 
x3 = v[2,2] * np.cos(np.sqrt(w[2])*t) 

fig1 = plt.figure()
ax = plt.axes()
ax.set_xlabel('t(s)')
plt.title('3C')
ax.plot(np.linspace(0,100,100),x1,color='blue', alpha=0.4)
ax.plot(np.linspace(0,100,100),x2,color='red', alpha=0.4)
ax.plot(np.linspace(0,100,100),x3,color='green', alpha=0.4)
```
